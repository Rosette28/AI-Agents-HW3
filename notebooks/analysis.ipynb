# System Performance Analysis

This notebook analyzes the CrewAI pipeline performance and API token costs.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
response_times = pd.DataFrame(
    {
        "RPM_10": [1.2, 1.8, 2.5],
        "RPM_30": [0.9, 1.3, 2.0],
        "RPM_60": [0.7, 1.1, 1.8],
    },
    index=["Small", "Medium", "Large"]
)

plt.figure(figsize=(8, 5))
sns.heatmap(response_times, annot=True)
plt.title("System Response Time Sensitivity")
plt.show()
```

# API Token Cost Analysis

```python
input_tokens = 2_135_000
output_tokens = 935_000

costs = pd.DataFrame(
    [
        {
            "Model": "GPT-4",
            "Input Tokens": input_tokens,
            "Output Tokens": output_tokens,
            "Estimated Cost ($)": (
                (input_tokens / 1_000_000) * 30
                + (output_tokens / 1_000_000) * 60
            ),
        },
        {
            "Model": "Claude 3",
            "Input Tokens": input_tokens,
            "Output Tokens": output_tokens,
            "Estimated Cost ($)": (
                (input_tokens / 1_000_000) * 15
                + (output_tokens / 1_000_000) * 75
            ),
        },
    ]
)

costs
```

```python
costs.plot(
    x="Model",
    y="Estimated Cost ($)",
    kind="bar",
    legend=False
)

plt.title("Estimated API Cost Comparison")
plt.show()
```
